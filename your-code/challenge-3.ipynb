```python
#Getting all the amount spent by each customer
#Ordering sum(amount_spent) by descending order
total_amount_customer=orders.pivot_table(index='CustomerID', values='amount_spent', aggfunc='sum').rename(columns={'amount_spent':'total_spent'})
total_amount_customer=total_amount_customer.sort_values(by=['total_spent'],ascending=False)
total_amount_customer.head(30)
```

```python
#Defining the percentiles
total_amount_customer=total_amount_customer.sort_values(by=['total_spent'],ascending=True)

q95=np.percentile(total_amount_customer['total_spent'],95)
print(round(q95,2))

q75=np.percentile(total_amount_customer['total_spent'],75)
print(q75)
```

```python
#merging the df 'orders' and the df 'total_amount_customer'

orders2 = pd.merge(left = orders, # LEFT TABLE
                                 right = total_amount_customer, # RIGHT TABLE
                                 how = 'inner', # TYPE OF JOIN
                                 left_on = "CustomerID", # COMMON COLUMN IN LEFT TABLE
                                 right_on= "CustomerID") # COMMON COLUMN IN RIGTH TABLE
display(orders2.shape)
orders2.head()
```

```python
orders2=orders2.rename(columns={"amount_spent_x": "amount_spent","amount_spent_y": "total_amount_spent"})
orders2.head()
```

```python
#adding a column called 'label' to identify vip and preferred customers

orders2['label']=np.nan

def get_customer_type(row):
    if row["total_spent"] > q95:
        return "VIP"
    elif row["total_spent"] > q75:
        return "Preferred"
    else:
        return "Other"

orders2["label"] = orders2.apply(get_customer_type, axis=1)
orders2.head()
 
```
